# Home Network Optimization Case Study
### AT&T Fiber + TP-Link Deco Mesh Architecture

**Author:** Richard Bernard  
**Date:** 2026-03-01  

---

## Overview
This case study documents the design, optimization, and performance tuning of a dual-node mesh network integrated with an ISP gateway using IP passthrough. The objective was to reduce latency and jitter for real-time workloads including cloud gaming and remote collaboration.


## Problem Statement

Observed issues included:
- Visual stutter during Amazon Luna gameplay
- Noticeable input latency under load
- Inconsistent roaming behavior between access points

Initial throughput tests showed >500 Mbps download speeds, indicating the issue was not bandwidth capacity but latency variance and RF instability.


## Architecture Before Optimization

- AT&T BGW320 Gateway (NAT + DHCP + Wi-Fi enabled)
- TP-Link Deco Mesh (Router mode, NAT enabled)
- Legacy Wi-Fi extender downstairs

**Result:** Double NAT topology and extender-based retransmission increased latency variance.


## Architecture After Optimization

- IP Passthrough enabled (DHCPS-fixed)
- AT&T Wi-Fi radios disabled
- Deco Mesh as single routing boundary
- 5 GHz @ 80 MHz channel width
- QoS configured using empirical speed testing (500 Mbps down / 350 Mbps up)

**Final Path:** Client → Deco → Public IP → ISP


## QoS Strategy

Real-world speeds were measured and slightly reduced in configuration to create headroom for queue management.

High Priority devices:
- Amazon Fire TV (Luna client)
- MacBook Pro (remote work / video conferencing)

Non-priority devices:
- Mobile phones
- Tablets
- Streaming-only endpoints


## Performance Validation

| Device        | Download (Mbps) | Upload (Mbps) | Latency (ms) |
|---------------|-----------------|---------------|--------------|
| MacBook Pro  | ~528–553       | ~346–382     | ~15          |
| iPhone       | ~316           | ~190         | ~17          |

Post-optimization results demonstrated stable low-latency performance and elimination of visible gameplay stutter.


## Key Engineering Decisions

- Eliminated double NAT to reduce routing complexity
- Replaced Wi-Fi extender with true mesh architecture
- Locked critical workloads to 5 GHz band
- Applied QoS shaping using measured throughput

This approach emphasizes architectural correctness over raw bandwidth.


## Takeaways

- High bandwidth does not guarantee low latency
- Extenders introduce retransmission overhead
- QoS must be based on measured speeds, not ISP marketing numbers
- Proper RF design and routing simplification dramatically improve real-time workloads
